# Data Analysis & Visualization

```python
df.describe()
```

```python
# --------------------------------------------
#  SECTION 1: DESCRIPTIVE ANALYSIS
# Goal: Understand the statistical overview of feedback data
# --------------------------------------------

# 1 Basic descriptive summary (mean, std, min, max, quartiles)
df.describe()

# 2 Calculate the mean rating for each parameter
mean_ratings = df.mean().round(2)
print(" Average Ratings per Parameter:\n", mean_ratings, "\n")

# 3 Find the highest-rated feedback aspect
highest_rated = mean_ratings.idxmax()
print(f" Highest Rated Aspect: {highest_rated} ({mean_ratings[highest_rated]})")

# 4 Find the lowest-rated feedback aspect
lowest_rated = mean_ratings.idxmin()
print(f" Lowest Rated Aspect: {lowest_rated} ({mean_ratings[lowest_rated]})")

# 5 Check which aspects have the most variation in responses
variation = df.std().round(2).sort_values(ascending=False)
print("\n Variation in Feedback (Standard Deviation):\n", variation)
```

# Visual Analysis (Exploratory Data Visualization)

## Bar Chart  Average Ratings per Parameter

```python
# Set a clean visual theme
sns.set(style="whitegrid", palette="viridis")
plt.figure(figsize=(10,6))

# 1 Average Ratings per Parameter
avg_ratings = df.mean().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=avg_ratings.values, y=avg_ratings.index, palette="crest")
plt.title("Average Ratings per Parameter", fontsize=14, fontweight='bold')
plt.xlabel("Average Rating")
plt.ylabel("Parameter")
plt.show()
```

```python
# 2 Box Plot  Variation in Ratings
plt.figure(figsize=(12,6))
sns.boxplot(data=df, palette="viridis")
plt.title("Variation in Ratings for Each Feedback Parameter", fontsize=14, fontweight='bold')
plt.xlabel("Parameter")
plt.ylabel("Rating Distribution")
plt.xticks(rotation=30)
plt.show()
```

```python
# 3 Correlation Heatmap
plt.figure(figsize=(10,6))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap="YlGnBu", linewidths=0.5)
plt.title("Correlation Heatmap of Feedback Parameters", fontsize=14, fontweight='bold')
plt.show()
```

```python
# 4 Distribution Plot (Histogram) for each rating parameter
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
plt.figure(figsize=(12,8))
for col in numeric_cols:
    sns.histplot(df[col], kde=True, label=col, bins=10)
plt.title("Distribution of Feedback Ratings", fontsize=14, fontweight='bold')
plt.xlabel("Rating")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
# 5 Overall Satisfaction Score (Optional)
df["Overall_Score"] = df.mean(axis=1)
plt.figure(figsize=(8,5))
sns.histplot(df["Overall_Score"], kde=True, color="skyblue", bins=10)
plt.title("Overall Student Satisfaction Distribution", fontsize=14, fontweight='bold')
plt.xlabel("Average Score per Student")
plt.ylabel("Number of Students")
plt.show()
```

```python
# Define satisfaction categories
def satisfaction_level(score):
    if score >= 7:
        return 'High Satisfaction '
    elif score >= 5:
        return 'Moderate Satisfaction '
    else:
        return 'Low Satisfaction '

# Apply the function
df['Satisfaction_Level'] = df['Overall_Score'].apply(satisfaction_level)

# Plot pie chart
plt.figure(figsize=(6,6))
df['Satisfaction_Level'].value_counts().plot.pie(autopct='%1.1f%%', colors=['#66b3ff','#99ff99','#ff9999'])
plt.title("Overall Satisfaction Levels", fontsize=14, fontweight='bold')
plt.ylabel("")
plt.show()
```
